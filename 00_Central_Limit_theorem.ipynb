Converting the 3d array of shape (768, 1024, 3) into a vector of shape (2359296)

```python
img1 = img1.flatten()
img2 = img2.flatten()
img3 = img3.flatten()
```

```python
img1.shape
```

Standardizing the image vectors

```python
img1 = (img1 - np.mean(img1))/np.std(img1)
img2 = (img2 - np.mean(img2))/np.std(img2)
img3 = (img3 - np.mean(img3))/np.std(img3)
```

Kurtosis tells about the shape of the tails of the distribution with reference to the shape of the whole distribution. The sharper image is going to have a high kurtosis value. 

```python
def kurtosis(img):
    k = np.mean(img**4) - 3
    return k
```

```python
images = [img1,img2,img3]
i =1
for img in images:
    k = kurtosis(img)
    print("Kurtosis for imgage",i,"is",k)
    i+=1
```

The sharpest image is image 1 which is the best out of the three images.
