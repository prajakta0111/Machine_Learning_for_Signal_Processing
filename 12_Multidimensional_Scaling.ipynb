### Part 2: Multidimensional Scaling

```python
dist =  scipy.io.loadmat(root + 'MDS_pdist.mat')
```

```python
dist = np.array(dist['L'])
```

```python
dist.shape
```

```python
def mean_s(M,ax):
    M = M - np.mean(M,axis=ax)
    return M
```

```python
M = mean_s(dist,1)
```

```python
M = mean_s(M,0)
```

```python
M.shape
```

```python
u,s,vh = np.linalg.svd(M)
```

```python
u.shape,s.shape,vh.shape
```

```python
diagonal = np.sqrt(np.diag(s))
```

```python
c = np.matmul(u,diagonal)
```
